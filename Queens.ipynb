## Problem Statement
The goal of this program is to arrange 8 queens on a board so they can't attack each other. 
There are several approaches that use heuristics.
A brute force method would look at all the possible solutions with 8 queens.
A chess board is an 8x8 grid.
56 positions will be empty. 8 will have queens.
There are multiple valid solutions. None are better than the others.

```python
# Lets make an empty board
# I know this is not efficient. I am spelling things out.
# A data frame would be better.
def new_board():
    board=[]
    for i in range(8):
        row =  [False for x in range(8)]
        board.append(row)
    return board

def print_board(board):
    # Assume board is 8x8 
    
    for i in range(16):
        if i % 2 == 0 :
            print(" " + "-" * 32)
        else:
            print("|", end= "")
            for j in range(8):
                if board[i//2][j]:
                    print(" Q ",end = "|")
                else:
                    print("   ", end = "|")
            print()
    print(" " + "-" * 32)
    
    
#Just for validation
test_board = new_board()
test_board[4][2]=True
test_board[7][7]=True
print_board(test_board)
```

#### My approach
I am going to take an approach of adding a Queen then checking constraints
If the constaints still hold true and we have fewer than 8 queens, add 1
If we have 8 queens and all good then done
If constaints are false, then back up
I will use a recursive alogorithm

```python
# We might need to increase the recursion limit
# See this article: https://careerkarma.com/blog/python-maximum-recursion-depth-exceeded-in-comparison/
# This is just a proof of concept, so I don't care if it works.
# Part of the demo is to show how this approach is bad

def check_attack(board,pos):
    # Only need to see if the new add causes an issue
    # We have checked everything else
    ok = True
    x = pos // 8
    y = pos % 8 
    
    queens = []
    for x_board in range(8):
        for y_board in range(8):
           
            if board[x_board][y_board]:
                # For testing
                queens.append((x_board,y_board))
                if  (x_board == x or y_board == y):
                    return False
                    # Look for anything with same x or y
                elif abs(x_board - x) == abs(y_board-y):
                    # On a diagnal
                    return False

    
    #print("Current Queens")
    #for queen in queens:
    #    print(queen)
    #print("I was checking", x,y)
    
    
    return ok  

    
# I am not sure if my laptop can handle the Naive approach
def add_queen(board,cur_queens,pos=0,debug=False):
    board_local = board.copy()
    if cur_queens == 0:
        board_local[0][0] = True # Lets assume 0,0 is a good place
        cur_queens = 1
        pos_local = 0
        temp_board = add_queen(board_local,1,pos_local)
        while (not temp_board) and pos_local < 64:
            pos_local += 1
            print("Check {} for Queen 2".format(pos_local))
            print_board(board_local)
            temp_board = add_queen(board_local,1,pos_local)
        return temp_board
    else:
        # Look for a valid place to put the next queen
        i = pos +1
        while i < 64:
            x = i // 8
            y = i % 8  
            if check_attack(board,i):
                print("Placing {},{} as queen{}".format(x,y,cur_queens+1))
                # This board is good, add another
                #print(i,"checked out")
                #print(cur_queens,"on the board")
                board_local[x][y]= True
                cur_queens +=1
                if cur_queens < 8:
                    temp_board = add_queen(board_local,cur_queens,i,debug)
                    if temp_board:
                        return temp_board
                    elif  i < 64:    # temp_board is false, but we have more on the current frontier
                        i=i+1 # check the next i
                        print("Failed but more to check")
                        board_local[x][y]= False
                        cur_queens -=1
                        continue
                    else: # we hit the end, back up
                        # It returned an empty board
                        print("Tried and failed")
                        cur_queens -=1
                        return  False 
                        # This will push the error back. 
                        #I could do this cleaner but I am writing this adhoc on a lot of coffee

                else: 
                    # We are up to 8 queens and are good. Send the board back down the recursion
                    return board
            else:
                i += 1
        print("Rewinding - Number of Queens",cur_queens)
        
        return False # We exhausted everything and didnt find a solution
               

        

board = new_board()
board = add_queen(board,0,True)
#print(board)
print_board(board)
```
